# 1.Write a program to print the sum of series 1 + 4 + 9 + 16 + 25 +  36 + .... for given N numbers.in python 

```python

n = int(input("Enter the value of N: "))

sum = 0

for i in range(1, n + 1):
    sum += i * i

print("Sum of the series is:", sum)
```

## 3.Write a program to enter the name, quantity, and price of five products by a user. Generate and print all the details with the product’s total amount in a formatted manner using a list.


```python
names = []
quantities = []
prices = []

for i in range(5):
    print("Enter details for product", i+1)
    
    name = input("Name: ")
    qty = int(input("Quantity: "))
    price = float(input("Price: "))
    
    names.append(name)
    quantities.append(qty)
    prices.append(price)

print("\nProduct\tQuantity\tPrice\tTotal")
print("------------------------------------------")

for i in range(5): 
    total = quantities[i] * prices[i]
    print(names[i], "\t", quantities[i], "\t\t", prices[i], "\t", total)
```

## 5)Write a program to find tuples with positive elements in the list of tuples.

```python
data = [(1, 2, 3), (-1, 2, 3), (4, 5, 6), (0, 2, 3)]

result = [i for i in data if all(x > 0 for x in i)]

print(result)
```

```python
data = [(1, 2, 3), (-1, 2, 3), (4, 5, 6), (0, 2, 3), (7, -2, 1)]

positive_tuples = []

for t in data:
    all_positive = True
    
    for num in t:
        if num <= 0:
            all_positive = False
            break
    
    if all_positive:
        positive_tuples.append(t)

print("Tuples with all positive elements:")
print(positive_tuples)
```

## 6) Write a program to sort a dictionary by key in ascending and descending order.

```python
n=int(input("Enter size:"))
d={}

for i in range(n):
    key=input("Enter key:")
    value=input("Enter value:")
    d[key]=value

asc=dict(sorted((d.items())))
print(asc)
desc=dict(sorted(d.items(),reverse=True))
print(desc)
```

## 7) Write a program to generate a Fibonacci series using a function.(e.g. 0 1 1 2 3 5 8..).

```python
n=int(input("Enter number:"))
def fibo(n):
    a=0
    b=1
    for i in range(1,n+1):
        print(a)
        c=a+b
        a=b
        b=c

fibo(n)
```

## 8) Write a program to enter a file name and check whether it exists or not. Ask for user confirmation and delete the file.

```python
import os

filename=input("Enter file name:")

if os.path.exists(filename):
    print("File exists")
    choice=input("Do you want to delete it? (yes/no): ")

    if choice.lower() == "yes":
        os.remove(filename)
        print("File deleted")
    else: 
        print("File not deleted")
else:
    print("File does not exists")
```
